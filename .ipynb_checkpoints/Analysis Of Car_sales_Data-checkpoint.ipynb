# Data Manipulating 

```python
carsales = pd.read_csv("https://raw.githubusercontent.com/mrdbourke/zero-to-mastery-ml/master/data/car-sales-missing-data.csv")
carsales
```

```python
carsales.dtypes
```

```python
carsales["Odometer"].mean()
```

```python
carsales["Odometer"] = carsales["Odometer"].fillna(carsales["Odometer"].mean())
```

```python
carsales
```

```python
carsales["Make"].fillna("BMW")
```

```python
carsales["Make"] = carsales["Make"].fillna("BMW")
```

```python
carsales
```

```python
carsales["Doors"] = carsales["Doors"].fillna(4.0)
```

```python
carsales
```

```python
carsales["Price"] = carsales["Price"].str.replace("[\$\,]", "").astype(float)
```

```python
carsales
```

```python
carsales["Price"].mean()
```

```python
carsales["Price"] = carsales["Price"].fillna(carsales["Price"].mean())
```

```python
carsales
```

```python
carsales["Price"] = carsales["Price"].astype(int)
```

```python
carsales
```

```python
carsales["Colour"] = carsales["Colour"].fillna("Green")
```

```python
carsales
```

```python
carsales.dtypes
```

```python
Engine_Size = [2.1, 4.5, 1.1, 5.6, 2.3, 3.3, 5.5, 4.1, 6.1, 3.0]
```

```python
carsales
```

```python
carsales
```

```python
Fuel_Km_Per_Hour = [4.5, 5.5, 7.6, 5.6, 4.2, 5.5, 6.7, 8.6, 5.5, 4.6]
```

```python
carsales["Fuel per 100KM"] = Fuel_Km_Per_Hour
```

```python
carsales
```

```python
seats = 5
```

```python
carsales["Seats"] = seats
```

```python
carsales["Seats"] = carsales["Seats"].fillna(5)
```

```python
carsales
```

```python
carsales["Total Used Fuel"] = carsales["Odometer"]/ 100 * carsales["Fuel per 100KM"]
```

```python
carsales["Total Used Fuel"]
```

```python
carsales
```

```python
carsales["Total Used Fuel"] = carsales["Total Used Fuel"].astype(int)
```

```python
carsales
```

```python
carsales["Trust & Safety"] = True
```

```python
carsales
```

```python
carsales["Odometer"] = carsales["Odometer"].apply(lambda x : x / 1.6)
```

```python
carsales
```

```python
carsales["Odometer"] = carsales["Odometer"].astype(int)
```

```python
carsales
```

```python
pd.crosstab(carsales["Colour"], carsales["Price"])
```

```python
carsales_suffle = carsales.sample(frac=1)
```

```python
carsales_suffle
```

```python
carsales_suffle = carsales_suffle.reset_index(drop = True)
```

```python
carsales_suffle
```
